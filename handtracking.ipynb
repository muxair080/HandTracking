```python
import cv2
import time
import mediapipe as mp
```

```python
cap = cv2.VideoCapture(0)
# mediapip predefine functions
mpHands = mp.solutions.hands
hands = mpHands.Hands()
# this object is showing points and join it on hand in front of camera
mpDraw = mp.solutions.drawing_utils
cTime = 0
pTime = 0
while True:
#     take image
    success, img = cap.read()
#     convert image to rgb.
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    results = hands.process(imgRGB)
#     print(results.multi_hand_landmarks)
    
    if results.multi_hand_landmarks:
#         loop is for multiple hands.
        for handlms in results.multi_hand_landmarks:
#            show landmards on hand in camera and join these points.
            mpDraw.draw_landmarks(img, handlms,
            mpHands.HAND_CONNECTIONS )
            
            """
            the value we are takking in x, y, z plain these value 
            are giving us the pixels or position but its in decimal
            or points we will multiple these values with the with and 
            height of image so we will get the exact pixels values
            of each point.
            """
            for id, lm in enumerate(handlms.landmark):
                h,w,c = img.shape
                cx, cy = int(lm.x*w), int(lm.y*h)
                print(id, cx, cy)
                if id ==0:
                    cv2.circle(img, (cx,cy), 25, 
                    (255,0,255), cv2.FILLED)
     
    cTime = time.time()
#     fps is frame per second
    fps = 1/(cTime-pTime)
    pTime  = cTime
#     show text on camera screen 
    cv2.putText(img, str(int(fps)), (10,70), cv2.FONT_HERSHEY_PLAIN,3
               ,(255,0,255,),3)
    cv2.imshow("Image", img)
    
    cv2.waitKey(1)
    
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Stop the camera
cap.release()
cv2.destroyAllWindows()
```
